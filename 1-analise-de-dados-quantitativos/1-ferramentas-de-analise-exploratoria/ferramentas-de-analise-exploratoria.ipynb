## Classificação de variáveis

- **Quantitativas**: quando assumem valores númericos;
- **Qualitativas**: quando seus possíveis valores não são númericos;

![Classificação das Variávies](imagens/classificacao-das-variaveis.png)

## Definições Básicas

- **População**: Conjunto de indivíduos ou objetos com pelo menos uma característica em comum;
- **Amostra**: Uma parte da população;
- **Dados Brutos**: Conjunto de dados que não tem uma ordem aparente;
- **Rol**: Conjunto de dados que tem um ordenamento, seja crescente ou decresente;
- **Amplitude Total**: Diferença entre o maior e o menor valor observado no conjunto de dados;
- **Frequência**: Quantidade de vezes que determinado dado se repete.

## Distribuições de Frequência

### Exemplo:

A distribuição de frequência a seguir representa as notas na disciplina de estatística numa turma de 40 alunos.

| Classe (Notas) | Frequência Absoluta | Frequência Relativa (%) |
| -------------- | ------------------- | ----------------------- |
| 0 a 2          | 1                   | 2,5                     |
| 2 a 4          | 5                   | 12,5                    |
| 4 a 6          | 12                  | 30                      |
| 6 a 8          | 15                  | 37,5                    |
| 8 a 10         | 7                   | 17,5                    |
| Soma           | 40                  | 100                     |

## Elementos da Distribuição de Frequência

### Limites de Classe

São as várias formas de expressar os limites de classe em uma distribuição de frequência. O limite à esquerda é chamado de limite inferior (Li) e o limite à direita é chamado de limite superior (Ls) da classe. Exemplos:
- Li |--| Ls: Fechado à esquerda, fechado à direita;
- Li |-- Ls (Mais comum): Fechado à esquerda, abero à direita;
- Li --| Ls: Aberto à esquerda, fechado à direita;

### Amplitute de Classes 

É a diferença entre o limite de superior e o limite inferior da classe: Ls - Li.

**Ponto Médio da Classe Xi**: É a média aritmética entre os limites, incluindo sempre ambos. Exemplo: Na classe 2 |-- 4 o ponto médio é (2 + 4)/ 2 = 3

### Frequência Relativa

100 * (Frequência Absoluta de determinda classe) / (Soma de todas as Frequências Absolutas)

### Frequência Acumulada

É o cúmulo das frequências absolutas. A partir da primeira frequência, somam-se as respectivas frequências absolutas.

Veja que, para a primeira classe, a frequência acumulada é igual à frequência absoluta. A partir daí, começamos a somar as frequências absolutas, de forma que a acumulada da segunda classe é a frequência absoluta da primeira classe mais a frequência absoluta da segunda classe, ou seja, F(2a classe )= F1 + F2 = 1 + 5 = 6 
Para determinar a frequência acumulada da terceira classe, somamos a frequência acumulada da segunda com a frequência absoluta da terceira, ou seja, F(3a classe ) = F(2a classe ) + F3 = 6 + 12 = 18, e assim por diante.

## Como construir uma distribuição de Frequência

- **Determinar o Número de Classes (k)**: k = $\sqrt{n}$, onde n é o tamanho do conjunto de dados ou da amostra. Obs. Geralmente n está entre 5 e 15 (Estatística Básica);
- **Determinar a Amplitude Total (AT)**: É a diferença entre o maior e o menor valor no conjunto de dados;
- **Determinar a Amplitude de Classe (AC)**: AC = AT / k

# [Representações Gráficas](https://estudante.estacio.br/disciplinas/estacio_13887132/conteudos/dfcc8d82-9bb9-45cd-a850-f521e11d2fab?tema=00520)

## Gráfico de barras ou colunas

É o tipo de gráfico mais utilizado em geral, pois serve para representar quaisquer dados quantitativos.
Exemplo: considere a distribuição de frequência referente à quantidade de famílias que receberam auxílio escolar por número de filhos.

```python
import matplotlib.pyplot as plt

numero_de_filhos = list(range(1,5))
quantidade_de_familias = [51, 39, 18, 12]

plt.bar(numero_de_filhos, quantidade_de_familias)
plt.title('Quantidade de Famílias por Número de Filhos')
plt.ylabel('Quantidade da Famílias')
plt.xlabel('Número de filhos')

plt.show()
```

## Histograma

É o gráfico típico da **distribuição de frequência**. A diferença desse gráfico para o gráfico de barras ou colunas
se dá pelo fato de as **colunas aprensentarem-se justapostas**, ou seja, sem espaçamento entre elas. Em geral, a abcissa desse gráfico é representada pelas classes e a ordenada é representada pela frequência absoluta ou relativa.

Exemplo: considere a distribuição de frequência a seguir, que representa as notas na disciplina de estatística em uma turma de 40 alunos

```python
import matplotlib.pyplot as plt

notas = [1.5, 3.0, 2.7, 4.5, 6.8, 7.2, 8.5, 9.0, 5.5, 4.2]   

# bins = classes
plt.hist(notas, bins=[0,2,4,6,8,10], edgecolor='black')

plt.title("Distribuição de Notas na Disciplina de Estatística")
plt.xlabel("Notas")
plt.ylabel("Quantidade de Alunos")

plt.show()
```

# Gráfico de Linhas

É o gráfico mais apropriado quando trabalhamos com uma **série de tempo**. Exemplo: número de acidentes por mês ao longo de uma ano:

```python
import matplotlib.pyplot as plt

meses = ["Jan", "Fev", "Mar", "Abr", "Mai", "Jun", "Jul", "Ago", "Set", "Out", "Nov", "Dez"]
numero_de_acidentes = [33, 22, 15, 11, 8, 15, 29, 20, 15, 18, 20, 25]

plt.plot(meses, numero_de_acidentes)

plt.title("Número de Acidentes")

plt.show()
```

# Setor

É o gráfico mais adequeado quando trabalhamos com porcentagens.

Exemplo: em uma pesquisa de satisfação sobre determinado produto, 55% dos entrevistados disseram que estavam satisfeitos, 35% disseram que estavam insatisfeitos e 10% disseram que eram indiferentes.

```python
import matplotlib.pyplot as plt

labels = ["Insatisfeitos", "Indiferentes", "Satisfeitos"]
sizes = [35, 10, 55]

plt.title("Pesquisa de Satisfação")
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['red', 'yellow', 'cyan'])

plt.show()
```
