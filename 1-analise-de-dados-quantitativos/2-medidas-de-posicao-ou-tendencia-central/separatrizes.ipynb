# Separatrizes

As separatrizes têm a função de dividir o conjunto de dados em certo número de partes iguais. **A mediana
que divide os dados em duas partes iguais é um caso particular de separatriz**. No entanto, outras 
separatrizes também têm papel de destaque na estatística como **quartis, decis e pecentis**.

Antes de introduzir as fórmulas, vamos pensar sobre o seguinte problema:

A corretora XYZ tem em seu portfólio 60 imóveis que foram distribuídos de acordo com seu valor de venda 
(em milhares de reais). Os 25% dos imóveis mais caros custam mais do que um valor V. Determine V.

| Classe (Notas) | Frequência Absoluta | Frequência Acumulada |
| -------------- | ------------------- | -------------------- |
| [0, 100[       | 5                   | 5                    |
| [100, 200[     | 17                  | 22                   |
| [200, 300[     | 20                  | 42                   |
| [300, 500[     | 13                  | 55                   |
| [500, 800[     | 5                   | 60                   |
| Total          | 60                  | -                    |

1) Os 25% dos imóveis mais caros custam mais do que um valor V implica que:

&rarr; Os 25% dos imóveis mais caros começam a partir da posição $0.25 * 60 = 45$;

&rarr; A posição 45 está na classe [300, 500[, que têm os valores das posições de 43 até 55;

&rarr; Na classe [300, 500[ temos 13 valores (55 - 43 + 1) e o 45 está na terceira posição dentro da classe (43,44,45).
Uma forma de chutar o valor do elemento 45 é por uma regra de 3, assumindo uma **distribuição
uniforme** dentro da classe:

$\frac{V_{E_{45}}}{200} = \frac{3}{13} \rightarrow V_{E_{45}} = 46.1538$

Conclusão: $V = 300 + 46.1538 = 346.15$

```python
import matplotlib.pyplot as plt

classes = [(0, 100), (100, 200), (200, 300), (300, 500), (500, 800)]
frequencias = [5, 17, 20, 13, 5]

larguras = [c[1] - c[0] for c in classes]
posicoes = [c[0] for c in classes]

# Gráfico
plt.figure(figsize=(8,5))
plt.bar(posicoes, frequencias, width=larguras, align='edge', edgecolor="black", alpha=0.7)
plt.xlabel("Valor do imóvel (milhares de R$)")
plt.ylabel("Frequência")
plt.title("Distribuição dos valores dos imóveis (agrupados)")

# Marcar o percentil 75 (V ≈ 346,15)
V = 346.15
plt.axvline(V, color="red", linestyle="--", linewidth=2, label=f"P75 ≈ {V:.2f}")

plt.legend()
plt.show()
```

# Formula Geral da Separatriz ($Q_n$)

$S_n = L_{i_{S_{n}}} + (\frac{E_{S_{n}} - F_{acc}}{F_{S_{n}}})*h$


Onde: 

- $S_n$ = Valor da separatriz;
- $L_{i_{S_{n}}}$ = Limite inferior da classe onde a separatriz se encontra;
- $E_{S_{n}}$ = Posição da separatriz;
- $F_{acc}$ = Frequência acumulada da classe onde a separatriz se encontra;
- $F_{S_{n}}$ = Frequência absoluta da classe onde a separatriz se encontra;

Aplicando a fórmula no problema anterior, temos:


$Q_n = 300 + (\frac{45 - 42}{13})*200 = 346.15$

# Decil

![Reta do Decil](images/decil.png)

Exemplo: Considere o exemplo da corretora, a partir de que valor estão os 10% dos imóveis mais baratos?

| Classe (Notas) | Frequência Absoluta | Frequência Acumulada |
| -------------- | ------------------- | -------------------- |
| [0, 100[       | 5                   | 5                    |
| [100, 200[     | 17                  | 22                   |
| [200, 300[     | 20                  | 42                   |
| [300, 500[     | 13                  | 55                   |
| [500, 800[     | 5                   | 60                   |
| Soma           | 60                  | -                    |

- $E_{D_{10}} = 0.1 * 60 = 6 \rightarrow$ o primeiro decil se encontra na classe [100, 200[, que
comporta os imóveis da posição 6 até 22;
- $h = 200 - 100 = 100$;
- $F_acc = 5$;
- $F_{D_{10}} = 22$;
- Utilizando a fórmula: $D_{10} = 100 + (\frac{6 - 5}{17})*100 = 105.88$

# Percentil

![Reta do Percentil](images/percentil.png)

Exemplo: Considere o exemplo da corretora, a partir de que valor estão 1% dos imóveis mais caros?

| Classe (Notas) | Frequência Absoluta | Frequência Acumulada |
| -------------- | ------------------- | -------------------- |
| [0, 100[       | 5                   | 5                    |
| [100, 200[     | 17                  | 22                   |
| [200, 300[     | 20                  | 42                   |
| [300, 500[     | 13                  | 55                   |
| [500, 800[     | 5                   | 60                   |
| Soma           | 60                  | -                    |

- $E_{P_{99}} = 0.99 * 60 = 59.4 \rightarrow$ o percentil 99 se encontra na classe [500, 800[;
- $h = 800 - 500 = 300$;
- $F_acc = 55$;
- $F_{P_{99}} = 5$;
- Utilizando a fórmula: $P_{99} = 500 + (\frac{59.4 - 55}{5})*300 = 764.00$
